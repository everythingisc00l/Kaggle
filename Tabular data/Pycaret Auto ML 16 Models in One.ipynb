# Pycaret. AutoML.

    In this notebook i'm loaded classical titanic dataset to pycaret - it's a instrument for autoML, where u can see training for your data in different classic algorthythms and then choose the best of them by metrics. Also this library provide excellent graphic visualization for EDA.

```python
data = pd.read_csv('/kaggle/input/titanic/train.csv')
data.head()
```

# 1.Use Pycaret for autoML :)

```python
from pycaret.classification import *
s = setup(data, target='Survived', session_id=123)
```

# 2. Find best models from 16 classic methods!!

```python
best = compare_models()
```

```python
print(best)
```

# Regression is best! :D

```python
evaluate_model(best)
```

```python
predict_model(best)
```

```python
test = pd.read_csv('/kaggle/input/titanic/test.csv')
test
```

```python
predictions = predict_model(best, data=test)
predictions.head()
```
